```python
import matplotlib.pyplot as plt

# Model names
models = ['Random Forest', 'RNN']

# Accuracy scores
accuracies = [61.8, 41.7]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['blue', 'red'])
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Comparison of Model Accuracies')
plt.ylim(0, 100)

# Adding text labels
for i in range(len(models)):
    plt.text(i, accuracies[i] + 1, f"{accuracies[i]}%", ha='center')

plt.show()
```

```python
import graphviz

# Create Digraph object with rankdir='LR' (left to right)
dot = graphviz.Digraph(graph_attr={'rankdir': 'LR'})

# Add nodes and edges with colors
dot.node('A', 'Gather Customer Data', style='filled', fillcolor='lightblue')
dot.node('B', 'EDA & Preprocessing of Data', style='filled', fillcolor='lightgreen')
dot.node('C', 'Applying K-means Algorithm & Visualizing Clusters', style='filled', fillcolor='lightyellow')
dot.node('D', 'Model Train & Test', style='filled', fillcolor='lightsalmon')
dot.node('E', 'Evaluation And Data Analysis', style='filled', fillcolor='lightpink')

dot.edges([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E')])

# Save and render the graph
dot.render('horizontal_flow_diagram', format='png', cleanup=True)

# Display the graph
dot
```

```python
import matplotlib.pyplot as plt

# Model names
models = ['Random Forest', 'RNN', 'LSTM']

# Accuracy scores
accuracies = [61.8, 43.2, 42.3]

# Set up the plot
plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracies, color=['blue', 'green', 'orange'])

# Add title and labels
plt.title('Model Accuracy Comparison', fontsize=24)
plt.xlabel('Models', fontsize=18)
plt.ylabel('Accuracy (%)', fontsize=18)

# Add accuracy values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval}%", ha='center', fontsize=14)

# Display the plot
plt.ylim(0, 100)  # Ensure the y-axis ranges from 0 to 100 for better comparison
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
